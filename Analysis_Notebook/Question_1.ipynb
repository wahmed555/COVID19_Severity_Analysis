### Sum of Deaths

```python
dfA_2=dfA_1[:]
dfA_2['Sum_Death_feb_March']=dfA_1.iloc[:,5:].sum(axis=1)
dfA_2
```

###  Death Rate Calculation

```python
dfA_3=dfA_2[:]
dfA_3['Death_Rate']=dfA_3['Sum_Death_feb_March']/60
dfA_3
```

### Sorting Contries/State with respect to Death Rate in Ascending order

```python
dfA_4=dfA_3[:]
dfA_5=dfA_4.sort_values(by="Death_Rate")
dfA_5
```
