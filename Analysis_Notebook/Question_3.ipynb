### Total Confirmed Cases Datasets

```python
Total_COVID_19_confirmed=dfC_2['Sum_Confirmed_feb_March'].sum(axis=0)
Total_COVID_19_confirmed
```

### Total Recovered Cases Datasets

```python
Total_COVID_19_Recoverd=dfB_2['Sum_Recovered_feb_March'].sum(axis=0)
Total_COVID_19_Recoverd
```

### Total Death Cases Datasets

```python
Total_COVID_19_Death=dfA_2['Sum_Death_feb_March'].sum(axis=0)
Total_COVID_19_Death
```

## Probability that a patient who has COVID will be recovered - P(recover)

```python
P_recover=Total_COVID_19_Recoverd/(Total_COVID_19_confirmed)
P_recover
```

## Probability that a patient who has COVID will Die - P(death)

```python
P_death=Total_COVID_19_Death/(Total_COVID_19_confirmed)
P_death
```
