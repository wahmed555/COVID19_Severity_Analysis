### Sum of Recovered Cases

```python
dfB_2=dfB_1[:]
dfB_2['Sum_Recovered_feb_March']=dfB_1.iloc[:,5:].sum(axis=1)
dfB_2
```

###  Recovery Rate Calculation

```python
dfB_3=dfB_2[:]
dfB_3['Recovery_Rate']=dfB_3['Sum_Recovered_feb_March']/60
dfB_3
```

### Sorting Contries/State with respect to Recovery Rate in Decending order

```python
dfB_4=dfB_3[:]
dfB_5=dfB_4.sort_values(by="Recovery_Rate", ascending=False)
dfB_5
```
